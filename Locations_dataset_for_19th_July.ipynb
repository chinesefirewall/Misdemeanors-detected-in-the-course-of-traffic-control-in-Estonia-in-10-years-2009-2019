```python
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime as dt
from dateutil import parser
```

```python
locations_19 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/locations_20190719.csv")
locations_19
#This is the Location dataset of bikes used on the 19th of july on the basis of latitude and longitude coordinates. A normal day, the next day to metallica concert
#So to compare with the metallica day data we will consider the same parameters as considered for 18th July dataset
```

```python
locations_19['route_code'] = locations_19['route_code'].astype('category')
locations_19['cyclenumber'] = locations_19['cyclenumber'].astype('category')
```

```python
locations_19['coord_date'] = pd.to_datetime(locations_19['coord_date'], format='%Y-%m-%d')
```

```python
locations_19["month"] = locations_19['coord_date'].dt.month_name().str.slice(stop=3)
```

```python
locations_19["day"] = locations_19['coord_date'].dt.day_name().str.slice(stop=3)
```

```python
locations_19["dates"] = locations_19['coord_date'].apply(lambda x: str(x)[8:10])
```

```python
locations_19['coord_time'] = locations_19['coord_time'].str.split("+", n = 1, expand = True)[0]
```

```python
locations_19['time'] = pd.to_datetime(locations_19['coord_time'], utc=True, errors='coerce')
```

```python
locations_19['hour'] = locations_19['time'].dt.hour
```

```python
locations_19 = locations_19.drop(['coord_date'], axis = 1)
```

```python
locations_19 = locations_19.drop(['time'], axis=1)
```

```python
locations_19
we have the same timeframe for this data as well.
#The concert timings were 16:00 to 23:00 on the day before i.e. 18th July and for the 18th july dataset we set the timeframe to be 14:00 to 02:00am of the next day. 
#So similarly for this dataset we took the timeframe to be from 19th 14:00 to 02:00am of 20th July. 
#Also, this dataset itself has values values from 00:00am of 20th July.
```

```python
filtered_loc19 = locations_19[((locations_19.day == 'Fri') & (locations_19.hour >= 14.0) | (locations_19.day== 'Sat') & (locations_19.hour <= 2.0))]
```

```python
filtered_loc19
#we now have the filtered data similar to the 18th July dataset after considering the same parameteres.
```

```python
filtered_loc19[["cyclenumber", "latitude", "longitude"]]
#splicing out non required columns, to create a csv suitable for the online tool: https://maps.co/
```
