```
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime as dt
from dateutil import parser
```

```
locations_26 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/locations_20190726.csv")
locations_26
#This is the Location dataset of bikes used on the 26th of july on the basis of latitude and longitude coordinates. A normal day, the date after an exact week to the 19th of July)
#So to compare with the metallica day(18th July) data we will consider the same parameters as considered for 18th, 19th and 25th July dataset.
```

```
locations_26['route_code'] = locations_26['route_code'].astype('category')
locations_26['cyclenumber'] = locations_26['cyclenumber'].astype('category')
```

```
locations_26['coord_date'] = pd.to_datetime(locations_26['coord_date'], format='%Y-%m-%d')
locations_26["month"] = locations_26['coord_date'].dt.month_name().str.slice(stop=3)
locations_26["day"] = locations_26['coord_date'].dt.day_name().str.slice(stop=3)
```

```
locations_26["dates"] = locations_26['coord_date'].apply(lambda x: str(x)[8:10])
locations_26['coord_time'] = locations_26['coord_time'].str.split("+", n = 1, expand = True)[0]
locations_26['time'] = pd.to_datetime(locations_26['coord_time'], utc=True, errors='coerce')
locations_26['hour'] = locations_26['time'].dt.hour
```

```
locations_26 = locations_26.drop(['coord_date'], axis = 1)
```

```
locations_26 = locations_26.drop(['time'], axis=1)
```

```
locations_26
#THe following steps are similar to the one did with 18th, 19th and 25th July datasets.
```

```
filtered_loc26 = locations_26[((locations_26.day == 'Fri') & (locations_26.hour >= 14.0) | (locations_26.day== 'Sat') & (locations_26.hour <= 2.0))]
```
