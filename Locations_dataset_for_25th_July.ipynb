```
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime as dt
from dateutil import parser
```

```
locations_25 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/locations_20190725.csv")
locations_25
#This is the Location dataset of bikes used on the 25th of july on the basis of latitude and longitude coordinates. A normal day, the date after an exact week to metallica concert(18th July)
#So to compare with the metallica day(18th July) data we will consider the same parameters as considered for 18th and 19th July dataset
```

```
locations_25['route_code'] = locations_25['route_code'].astype('category')
locations_25['cyclenumber'] = locations_25['cyclenumber'].astype('category')
```

```
locations_25['coord_date'] = pd.to_datetime(locations_25['coord_date'], format='%Y-%m-%d')
locations_25["month"] = locations_25['coord_date'].dt.month_name().str.slice(stop=3)
locations_25["day"] = locations_25['coord_date'].dt.day_name().str.slice(stop=3)
```

```
locations_25["dates"] = locations_25['coord_date'].apply(lambda x: str(x)[8:10])
locations_25['coord_time'] = locations_25['coord_time'].str.split("+", n = 1, expand = True)[0]
locations_25['time'] = pd.to_datetime(locations_25['coord_time'], utc=True, errors='coerce')
locations_25['hour'] = locations_25['time'].dt.hour

```

```
locations_25 = locations_25.drop(['coord_date'], axis = 1)
```

```
locations_25 = locations_25.drop(['time'], axis=1)
```

```
locations_25
#The next steps are similar to the 18th and 19th July datasets since we considered the same parameters.
```

```
filtered_loc25 = locations_25[((locations_25.day == 'Thu') & (locations_25.hour >= 14.0) | (locations_25.day== 'Fri') & (locations_25.hour <= 2.0))]
```
