```python
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime as dt
from dateutil import parser
```

```python
locations_18 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/locations_20190718.csv")
locations_18


#This is the Location dataset of bikes used on the 18th of july on the basis of latitude and longitude coordinates. The day of Metallica Concert. 
#We need Locations of these bikes during the time of concert.
#The concert timings were 16:00 to 23:00. So we need only the bike coordinates from 14:00 to 02:00 am of the next day.
```

```python
locations_18['route_code'] = locations_18['route_code'].astype('category')
locations_18['cyclenumber'] = locations_18['cyclenumber'].astype('category')
```

```python
locations_18['coord_date'] = pd.to_datetime(locations_18['coord_date'], format='%Y-%m-%d')
```

```python
locations_18["month"] = locations_18['coord_date'].dt.month_name().str.slice(stop=3)
```

```python
locations_18["day"] = locations_18['coord_date'].dt.day_name().str.slice(stop=3)
```

```python
locations_18["dates"] = locations_18['coord_date'].apply(lambda x: str(x)[8:10])
```

```python
locations_18['coord_time'] = locations_18['coord_time'].str.split("+", n = 1, expand = True)[0]
```

```python
locations_18['time'] = pd.to_datetime(locations_18['coord_time'], utc=True, errors='coerce')
```

```python
locations_18['hour'] = locations_18['time'].dt.hour
```

```python
locations_18 = locations_18.drop(['coord_date'], axis = 1)
```

```python
locations_18 = locations_18.drop(['time'], axis=1)
```

```python
locations_18
#The datastet not only has the coordinates for just 18th july, but a also coordinates from 0.00 am to 2:00 am of 19th July as well.
```

```python
filtered_loc18 = locations_18[((locations_18.day == 'Thu') & (locations_18.hour >= 14.0) | (locations_18.day== 'Fri') & (locations_18.hour <= 2.0))]
```

```python
filtered_loc18
#So now we have the filtered data according to our requirements.
```

```python
filtered_loc18[["cyclenumber", "latitude", "longitude"]]
#We will now use these coordinate values to implement into a heatmap using an online tool: https://maps.co/
#But for that we only require latitude and longitude values. So we splice out the non required columns.
```

```python
coords18 = filtered_loc18[["cyclenumber", "latitude", "longitude"]]
```
